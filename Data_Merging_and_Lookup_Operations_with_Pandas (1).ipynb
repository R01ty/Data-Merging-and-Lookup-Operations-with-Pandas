```python
import pandas as pd
```

```python
TASK 1: Data Preparation and Missing Value Handling
```

```python
Step 1: Create DataFrames
```

```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)

print("Original Students DataFrame")
display(students)
```

```python
Step 2: Null Value Analysis
```

```python
print("Null Value Analysis")

for column in students.columns:
    null_count = students[column].isnull().sum()
    percentage = (null_count / len(students)) * 100
    print(f"Column: {column}, Nulls: {null_count} ({percentage:.2f}%)")
```

```python
| Column     | Nulls |
| ---------- | ----- |
| student_id | 0     |
| name       | 1     |
| email      | 1     |
| city       | 1     |
```

```python
Step 3: Fill Missing City Values
```

```python
students['city'] = students['city'].fillna('Unknown')
```

```python
Step 4: Drop Rows where Name is Missing
```

```python
students_clean = students.dropna(subset=['name'])
```

```python
Step 5: Display Cleaned DataFrame
```

```python
print("Cleaned Students DataFrame")
display(students_clean)
```

```python
Final rows = 6 students (student 103 removed).
```

```python
TASK 3: Lookup Operation and Automation
```

```python
1️⃣ Lookup using Dictionary + map()
```

```python
Create Dictionary
```

```python
score_dict = dict(zip(scores['student_id'], scores['exam_score']))
print(score_dict)
```

```python
Example:

{101:85, 102:92, 104:78, 105:88, 106:95}
```

```python
Add Scores using map()
```

```python
students_clean['exam_score'] = students_clean['student_id'].map(score_dict)

display(students_clean[['student_id','name','exam_score']])
```

```python
Output:

| student_id | name  | exam_score |
| ---------- | ----- | ---------- |
| 101        | Alice | 85         |
| 102        | Bob   | 92         |
| 104        | David | 78         |
| 105        | Emma  | 88         |
| 106        | Frank | 95         |
| 107        | Grace | NaN        |

Grace has NaN because no score exists.
```

```python
Why map() is More Efficient

Lookup using map() is faster because:

1️⃣ It performs direct dictionary lookups (O(1)).
2️⃣ No need to create a new DataFrame structure.
3️⃣ Lower memory usage.
4️⃣ Best when adding a single column from key-value mapping.

merge() is better when:

Joining multiple columns

Combining multiple tables
```

```python
3️⃣ Automation Function
```

```python
def auto_merge(df1, df2, join_type, key_column):

    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)

    result = {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

    return result
```
